```python
import time

def resumidor_emails(lista_de_emails):
  for corpo_email in lista_de_emails:
    time.sleep(25)
    response = client.models.generate_content(
      model='gemini-2.5-flash',
      contents=f'''Vou te enviar o corpo de um emai. Quero que você resuma em apenas 1 linha,
                passando o intuito daquele e-mail. Segue o email: {corpo_email}'''
      )
    print(response.text)
    print('-' * 50)
```

```python
resumidor_emails(corpos_emails)
```

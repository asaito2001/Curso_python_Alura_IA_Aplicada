```python
import pandas as pd
import numpy as np

# Gerar 50 linhas de dados para cada coluna
num_rows = 50

# Coluna 'produto'
produtos = [
    'Notebook', 'Smartphone', 'Smart TV', 'Fones de Ouvido', 'Teclado Mecânico',
    'Mouse Gamer', 'Monitor UltraWide', 'Câmera Digital', 'Console de Videogame', 'Tablet',
    'Impressora', 'Roteador Wi-Fi', 'Caixa de Som Bluetooth', 'Webcam HD', 'Microfone USB',
    'Disco Rígido Externo', 'SSD Interno', 'Placa de Vídeo', 'Memória RAM', 'Processador',
    'Fonte de Alimentação', 'Gabinete Gamer', 'Placa Mãe', 'Cooler para CPU', 'Mouse Pad',
    'Mochila para Notebook', 'Carregador Portátil', 'Smartwatch', 'Bateria Externa', 'Cabo HDMI',
    'Adaptador USB-C', 'Controle de Videogame', 'Fone Bluetooth', 'Drone', 'Cadeira Gamer',
    'Óculos de Realidade Virtual', 'Tripé para Câmera', 'Anel de Luz LED', 'Lâmpada Inteligente', 'Tomada Inteligente',
    'Sensor de Porta/Janela', 'Fechadura Eletrônica', 'Termostato Inteligente', 'Assistente Virtual', 'Fone com Fio',
    'Mouse Sem Fio', 'Teclado Sem Fio', 'Pen Drive', 'Cartão de Memória', 'Leitor de Cartões'
]
produto_data = np.random.choice(produtos, num_rows, replace=True)

# Coluna 'categoria do produto'
categorias = [
    'Eletrônicos', 'Informática', 'Acessórios', 'Games', 'Automação Residencial',
    'Áudio', 'Fotografia', 'Periféricos', 'Armazenamento', 'Componentes'
]
categoria_data = np.random.choice(categorias, num_rows, replace=True)

# Coluna 'preço do produto'
preco_data = np.round(np.random.uniform(50.00, 5000.00, num_rows), 2)

# Coluna 'avaliação do produto'
avaliacao_data = np.round(np.random.uniform(1.0, 5.0, num_rows), 1)

# Coluna 'quantidade vendida'
quantidade_data = np.random.randint(1, 1000, num_rows)

# Criar o DataFrame
df = pd.DataFrame({
    'produto': produto_data,
    'categoria_do_produto': categoria_data,
    'preco_do_produto': preco_data,
    'avaliacao_do_produto': avaliacao_data,
    'quantidade_vendida': quantidade_data
})

# Exibir as primeiras 5 linhas do DataFrame
display(df.head())
```

1. Carregar um arquivo .csv com feedback de clientes (reviews.csv)

3. Adicionar essa classificação ao Dataframe.

```python
import os
from google.colab import userdata

os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')
```

```python
from groq import Groq

client = Groq()
```

```python
def resposta_sentimento_llm(review, model_llm='llama-3.3-70b-versatile'):

  response = client.chat.completions.create(
      model=model_llm,
      messages=[{'role': 'user', 'content': review}],
      temperature=0,
      max_completion_tokens=20,
      stream=False
  )

  return response.choices[0].message.content.strip()
```

### Few-shot-prompting

```python
def classificar_sentimento(texto):

    prompt = f"""
Classifique o sentimento do texto abaixo.

Categorias possíveis:
Positivo, Negativo ou Neutro.

Exemplos:
'Eu adorei esse produto' -> 'Positivo'
'Gostei, mas não é nada de especial' -> 'Neutro'
'Odiei esse produto' -> 'Negativo'

Review a ser analisado: {texto}

Responda apenas com uma palavra.
"""

    try:
        resposta = resposta_sentimento_llm(prompt)

        if resposta not in ["Positivo", "Negativo", "Neutro"]:
            return "Indefinido"

        return resposta

    except Exception:
        return "Erro"
```

```python
import pandas as pd

df = pd.read_csv('/content/reviews.csv')
df_review = df['reviewText']
```

```python
lista_review_sentimento = []

for review_indice, review_texto in enumerate(df_review):
  sentimento = classificar_sentimento(review_texto)
  lista_review_sentimento.append(sentimento)

  #print(f'Idx review: {review_indice} | review: {review_texto} -> {sentimento}')
```

```python
df['Sentimento'] = lista_review_sentimento
```

```python
df.to_csv(
    "reviews_com_sentimentos.csv",
    index=False,
    sep=",",
    encoding="utf-8-sig"   # evita problema com acentos
)

print("CSV gerado com sucesso.")
```

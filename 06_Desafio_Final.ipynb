# Tratamento de erros com Try, Except e Finally

```python
try:
  int('abc')
except ValueError:
  print('número invalido!')
```

```python
# Capturar o erro do ValueError
try:
  int('abc')
except ValueError as ve:
  print('número invalido!')
  print('Detalhes do erro', ve)
```

```python
# Capturar diferentes tipos de erros
try:
  pessoas = 6
  pessoas_2 = 'Mario, Joao, Silvia, Maria'
  pessoas + pessoas_2
except ValueError as ve:
  print('Nao e um inteiro invalido!')
except TypeError as te:
  print('Nao somar um inteiro com uma string')
  print('Detalhe erro: ', te)
except Exception as e:
  print('Detalhe erro: ', e)
finally:
  print('Processo finalizado!')
```

# Categorização de reviews negativos

```python
import os
from google.colab import userdata

os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')
```

```python
from groq import Groq

client = Groq()
```

```python
def categorizacao_llm(review, model_llm='llama-3.3-70b-versatile'):

  response = client.chat.completions.create(
      model=model_llm,
      messages=[{'role': 'user', 'content': review}],
      temperature=0,
      max_completion_tokens=100,
      stream=False
  )

  return response.choices[0].message.content.strip()
```

```python
def categorizar_reviews(texto):

    prompt = f"""
Você é um analista de reviews.

Analise os reviews negativos abaixo e identifique
5 categorias principais de reclamações.

Regras:
- cada categoria deve ter apenas uma palavra
- letras minúsculas
- sem acentos
- separar por vírgula

Exemplo:
entrega, qualidade, preco, embalagem, atendimento

Reviews:
{texto}
"""
    resposta = categorizacao_llm(prompt)

    return resposta
```

```python
categorias = categorizar_reviews(reviews_negativos_union)
lista_de_categorias = categorias.split(sep=", ")
lista_de_categorias
```
